This notebook goes over the different methods of loading images with matplotlib and PIL.
It also converts them to grayscale images

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.image as mpimg
```

```python
#Loading image using matplotlib
img = mpimg.imread('pics/flower.png')
print(f'The shape is {img.shape} \n\n')
print(img)
```

```python
imgplot = plt.imshow(img)
```

```python
# applying grayscale filter

R, G, B = img[:,:,0], img[:,:,1], img[:,:,2]
imgGray = 0.2989 * R + 0.5870 * G + 0.1140 * B
plt.imshow(imgGray, cmap='gray')
plt.show()
```

```python
# opening an image with pillow instead
from PIL import Image, ImageOps
from IPython.display import display

pil_image = Image.open("pics/flower.png")
scale = 2.0
display(pil_image.resize(( int(pil_image.width * scale), int(pil_image.height * scale))))
```

```python
# converting pil img into numpy array
pil_nparray = np.array(pil_image)
```

```python
print(pil_nparray.shape)
print(pil_nparray.dtype)
print(f'Here are the values {pil_nparray}')
```

```python
gray_pil_im = ImageOps.grayscale(pil_image)
display(gray_pil_im)
```

```python
# resizing image to 28x28 pixels
gray_pil_im = gray_pil_im.resize(size=(224,224))
```

```python
display((gray_pil_im))
plt.figure(figsize=(6, 6))
plt.imshow(gray_pil_im, cmap='gray')
```
